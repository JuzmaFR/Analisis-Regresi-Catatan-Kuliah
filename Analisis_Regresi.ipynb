###Penafsiran Modern Regresi

Analisis Regresi berkenan dengan studi **ketergantungan** satu variabel, variabel tak bebas (*response variable*, variabel Y), pada satu atau lebih variabel lain, variabel yang menjelaskan(*explanatory variable*, variabel x)

### Tujuan

1. Prediksi
- Waktu
- TIdak dengan waktu

2. Seleksi Variabel, Prediksi akan akurat ketika variabel relevan

3. Spesifikasi Model, Apakah lebih baik menggunakan linear, quadratic, etc

4. Estimasi Parameter

### Stokhastik dan Deterministik

Regresi = Ketergantungan Statistik = Stokhastik

$Y = X\beta + \epsilon$


Regresi $\ne$ Ketergantungan Fungsional = Deterministik

Contoh

Fungsi Linear, Deterministik

$Y = X$

ketika X = 1, maka Y **pasti** 1 (Y = 1). Ketika X = 2, Y = 2. Dan seterusnya.

Regresi

$Y = \beta X + \epsilon$

misalnya X adalah Pendapatan dan Y adalah Konsumsi. Orang yang mempunyai pendapatan 5 juta per bulan belum tentu memiliki konsumsi yang sama. Ada orang yang lebih hedon terbiasa makan di mcd, sementara ada pula yang sehari-hari makan di warteg. Sehingga nilai Y berbeda-beda pada X tertentu. Oleh karena itu terdapat error ($\epsilon$) dan regresi bersifat stokhastik.

Suatu hubungan statistik, bagaimanapun kuat dan sugestif, tidak pernah dapat menetapkan hubungan sebab akibat - Kendal and Stuart

*Causation* $\rightarrow$ hal menyebabkan $\rightarrow$ hubungan sebab akibat.

1. Korelasi yang kuat antara variabel

2. Timer Order

Mana Variabel yang dapat diketahui lebih dahulu. X digunakan terlebih dahulu yang nantinya akan memprediksi Y.



3. Dukungan Teori

Model Regresi Dugaan

## Multikolinearitas

### Pendeteksian Multikolinearitas

####Determinan

yaitu dengan melihat nilai determinan dari matriks korelasi variabel x. Apabila nilainya mendekati 0, maka terdapat multikolinearitas. Sebaliknya jika nilai determinan mendekati 1, maka tidak terdapat multikolinearitas.

```r
#Membuat Matriks X
x <- as.matrix(longley[,-7]) #[,-7] untuk mengambil variabel selain Employed
x
```

```r
det(cor(x));
```

Dari hasil di atas dapat diketahui bahwa nilai determinan dari matriks korelasi X yaitu 1.57961548624822e-08 dimana nilai tersebut cenderung mendekati 0 sehingga dapat dikatakan bahwa terdapat multikolinearitas.

#### Indeks Kondisi
$ID = \sqrt{K}$ dengan $K = \frac{\lambda_{max}}{\lambda_{min}}, \lambda$: Nilai Eigen $X^{*t}X^*$ <br>
Dengan Kriteria: <br>
$100 \leq K \leq 1000$, multikolinearitas sedang menuju kuat. <br>
$K > 1000$, multikolinearitas sangat kuat. <br> <br>
$ID < 10$, multikolinearitas lemah. <br>
$10 \leq ID \leq 30$, multikolinearitas sedang menuju kuat. <br>
$ID > 30$, multikolinearitas sangat kuat.


```r
eigen_vector <- eigen(cor(x))

K <- max(eigen_vector$values)/min(eigen_vector$values)
K
```

Nilai 12220.0098602747 menandakan bahwa $K > 1000$ yang artinya multikolinearitas sangat kuat.

```r
sqrt(K);
```

110.5441534423 juga menandakan bahwa $ID > 30$ yang artinya multikolinearitas sangat kuat

#### Variance Inflation Factor (VIF)

$VIF = \frac{1}{1-R^2_{xx}}$, dengan kriteria jika VIF > 10, maka terdapat indikasi multikolinearitas. Secara singkat VIF ini mengambil nilai diagonal dari matriks invers korelasi. Nantinya secara berurut (sesuai matriks korelasi) dapat diambil nilai VIF dari masing-masing variabel.

```r
xinv = solve(cor(x))
diag(xinv)
```

Nilai VIF nya cenderung lebih besar dari 10. Hanya 1 yang tidak yaitu Armed.Forces dengan VIF 3.58893 yang sebelumnya sudah terbukti dengan melihat signifikansi.

### Penanggulangan Multikolinearitas

####Principal Component Regression

image.png

Metode Principal Component Regression yaitu metode dengan membuat masing-masing variabel X yang awalnya berkorelasi (tumpang tindih) di transformasikan ke dalam bentuk PC yang masing-masing PC nya nanti memuat seluruh informasi dari setiap variabel X. Tujuan dari PCR ini yaitu untuk mencapai **Orthogonalitas yaitu bagaimana membuat set faktor atau komponen dari variabel X1, X2, ... , Xn saling orthogonal** yaitu nilai dari $cor(K_1,K_2) = 0 $ (K adalah ... ).

##### Metode Pengambilan PC

image.png

image.png

1. Dengan $\lambda$ Values

Metode pengambilan dengan $\lambda$ dilakukan dengan melihat apakah $\lambda$ melebihi 1 atau belum ($\lambda > 1$). Jika sudah melebihi maka pengambilan PC dihentikan.

2. Dengan Proporsi

Metode ini dilakukan dengan melihat besaran proporsi dari $\lambda$ values nya. Penentuan kriteria proporsi biasanya secara subjektif. Namun umumnya lebih dari 80%.



Notes:

metode pertama dan kedua berhubungan. Jika pengambilan $\lambda <1$, setidaknya proporsi sudah lebih dari 80%. Sebaliknya jika $\lambda > 1$, setidaknya proporsi sudah lebih dari 70%.

Sehingga jika kita ambil contoh di atas, Pengambilan dapat di ambil PC1 sampai PC2 dengan $\lambda = 1.48049 \rightarrow 1.48049>1$ dan Proporsi Kumulatif = $ 0.794 = 79.4% \rightarrow 79.4$%$>70$ %. 

Selain itu juga bisa di ambil PC1 sampai PC3 dengan $\lambda = 0.64810297 \rightarrow 0.64810297<1$ dan Proporsi Kumulatif = $0.924 \rightarrow 92.4$ % $>80$ %.

Pengambilan PC berdasarkan kondisi di atas subjektif (???).

3. Scree Plot

Metode ini digunakan dengan melihat plot antara $PC( PC1 = 1, PC2 = 2, PC3 = 3, ..., PCn = n)$ dengan nilai $\lambda$. Pengambilan dihentikan ketika terjadi perubahan yang curam.

2. Merubah Zy dan Zx

Kita ketahui bahwa rumus dari Z secara umum yaitu:

$Z = \frac{x-\bar{x}}{\sigma}$

sehingga rumus dari Zy yaitu:

$Zy = \frac{Y- \bar{Y}}{Sy}$

$Sy$ merupakan simpangan baku dari Y.  sehingga $Y = ZSy+\bar{Y}$

Begitupun dengan Zx yaitu:

$Zx =\frac{X-\bar{x}}{Sx}$

Sx merupakan simpangan baku dari X. Sehingga $X = ZSx + \bar{X}$

Merubah Zx

### Dampak Terjadinya Heteroskedastisitas

1. Metode OLS masih tak bias namun tidak efisien $\rightarrow$ varians tidak minimum $\rightarrow$ berpengaruh pada efisiensi

2. Peramalan nilai variabel tak bebas Y berdasarkan model regresi yang dibangun tidak efisiean

### Konsekuensi Heteroskedastisitas

Varians taksirannya tidak minimum dan cenderung *under estimate* (Varians taksiran yang kecil). Minimum dalam konteks ini yaitu batas bawahnya ... bukan nilainya paling kecil $\rightarrow$ *under estimate* dan *over estimate*. Konsekuensinya saat melakukan *t-stundent* tanpa penanggulangan heteroskedastisitas nilai *t-student* akan cenderung besar. Sehingga bisa saja suatu pengujian hipotesis memiliki kesimpulan yang signifikan dengan *t-student* yang besar, padahal hal tersebut dikarenakan adanya heteroskedastisitas yang menyebabkan pengujian menjadi signifikan. Oleh karena itu, **perlu dilakukan terlebih dahulu Pengujian Asumsi kemudian lanjut ke Pengujian Hipotesis.**

$Y_i = \beta_0 + \beta_1x_i + \epsilon_i$

Misalkan:

$E(\epsilon_i^2) = \sigma^2$

Dengan

$\sum_{i=1}^n w_ie_i^2 = \sum_{i=1}^n w_i(Y_i - \beta_0X_i)$ 

$w_i = \frac{1}{\sigma_i}$

adanya *weight*($w_i$) atau pembobotan diperlukan untuk mendapatkan varians dari $e_i$ yang konstan.

$Var(e_i) \approx 1 \rightarrow Var(\epsilon) \approx \sigma^2$

Jika varians dari $e_i$ konstan, maka dapat di asumskan varians error nya mendekati $\sigma^2$.

sehingga $\hat{\beta_1}$ dari :

$\hat{\beta_1} = \frac{\sum X_iY_i - \sum X_i \sum Y_i}{\sum X_i^2 - (\sum X_i)^2}$

menjadi

$\hat{\beta_1} = \frac{(\sum w_i) (\sum w_i\sum X_iY_i) - (\sum w_i\sum X_i) (\sum w_i \sum Y_i)}{(\sum w_i)(\sum w_i\sum X_i^2) - (\sum w_i\sum X_i)^2}$ 

Var$(\hat{\beta_1}) = \frac{\sum w_i}{(\sum w_i)(\sum w_iX_i^2) - (\sum w_iX_i)^2}$

Contoh

$Y_i = \beta_0 + \beta_1X_1 + \epsilon_i$

$\epsilon_i = \beta_2X_2 + V_i$

Ini dapat menyebabkan heteroskedastisitas karena $\epsilon$ memiliki $X_2$ yang berkaitan dengan $Y_i$ sehingga diperlukan penanggulangan heteroskedastisitas. Jadi Heteroskedastisitas terjadi ketika terdapat variabel x yang berada dalam $\epsilon$.

Jadi $V(\epsilon_i)$ terhadap heteroskedastisitas dipengaruhi oleh X karena $V(\epsilon_i) = \sigma^2_i$ sebenarnya tidak terlepas dari sisa yaitu $\epsilon_i = \{ X_n \}$
