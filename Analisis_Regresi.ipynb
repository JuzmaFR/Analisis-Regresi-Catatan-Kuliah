## Multikolinearitas

### Pendeteksian Multikolinearitas

####Determinan

yaitu dengan melihat nilai determinan dari matriks korelasi variabel x. Apabila nilainya mendekati 0, maka terdapat multikolinearitas. Sebaliknya jika nilai determinan mendekati 1, maka tidak terdapat multikolinearitas.

```r
#Membuat Matriks X
x <- as.matrix(longley[,-7]) #[,-7] untuk mengambil variabel selain Employed
x
```

```r
det(cor(x));
```

Dari hasil di atas dapat diketahui bahwa nilai determinan dari matriks korelasi X yaitu 1.57961548624822e-08 dimana nilai tersebut cenderung mendekati 0 sehingga dapat dikatakan bahwa terdapat multikolinearitas.

#### Indeks Kondisi
$ID = \sqrt{K}$ dengan $K = \frac{\lambda_{max}}{\lambda_{min}}, \lambda$: Nilai Eigen $X^{*t}X^*$ <br>
Dengan Kriteria: <br>
$100 \leq K \leq 1000$, multikolinearitas sedang menuju kuat. <br>
$K > 1000$, multikolinearitas sangat kuat. <br> <br>
$ID < 10$, multikolinearitas lemah. <br>
$10 \leq ID \leq 30$, multikolinearitas sedang menuju kuat. <br>
$ID > 30$, multikolinearitas sangat kuat.


```r
eigen_vector <- eigen(cor(x))

K <- max(eigen_vector$values)/min(eigen_vector$values)
K
```

Nilai 12220.0098602747 menandakan bahwa $K > 1000$ yang artinya multikolinearitas sangat kuat.

```r
sqrt(K);
```

110.5441534423 juga menandakan bahwa $ID > 30$ yang artinya multikolinearitas sangat kuat

#### Variance Inflation Factor (VIF)

$VIF = \frac{1}{1-R^2_{xx}}$, dengan kriteria jika VIF > 10, maka terdapat indikasi multikolinearitas. Secara singkat VIF ini mengambil nilai diagonal dari matriks invers korelasi. Nantinya secara berurut (sesuai matriks korelasi) dapat diambil nilai VIF dari masing-masing variabel.

```r
xinv = solve(cor(x))
diag(xinv)
```

Nilai VIF nya cenderung lebih besar dari 10. Hanya 1 yang tidak yaitu Armed.Forces dengan VIF 3.58893 yang sebelumnya sudah terbukti dengan melihat signifikansi.

Metode Principal Component Regression yaitu metode dengan membuat masing-masing variabel X yang awalnya berkorelasi (tumpang tindih) di transformasikan ke dalam bentuk PC yang masing-masing PC nya nanti memuat seluruh informasi dari setiap variabel X. ### Dampak Terjadinya Heteroskedastisitas

1. Metode OLS masih tak bias namun tidak efisien $\rightarrow$ varians tidak minimum.

2. Peramalan nilai variabel tak bebas Y berdasarkan model regresi yang dibangun tidak efisiean

3. Selang kepercayaan semakin tidak efisien $\rightarrow$ sempit.

4. Pengujian hipotesis signifikansi parameter model regresi baik dengan uji signifikansi maupun pendekatan selang kepercaayaan tidak dapat diterapkan.

### Konsekuensi Heteroskedastisitas

Varians taksirannya tidak minimum dan cenderung *under estimate* (Varians taksiran yang kecil). Minimum dalam konteks ini yaitu batas bawahnya ... bukan nilainya paling kecil $\rightarrow$ *under estimate* dan *over estimate*. Konsekuensinya saat melakukan *t-stundent* tanpa penanggulangan heteroskedastisitas nilai *t-student* akan cenderung besar. Sehingga bisa saja suatu pengujian hipotesis memiliki kesimpulan yang signifikan dengan *t-student* yang besar, padahal hal tersebut dikarenakan adanya heteroskedastisitas yang menyebabkan pengujian menjadi signifikan. Oleh karena itu, **perlu dilakukan terlebih dahulu Pengujian Asumsi kemudian lanjut ke Pengujian Hipotesis.**

$Y_i = \beta_0 + \beta_1x_i + \epsilon_i$

Misalkan:

$E(\epsilon_i^2) = \sigma^2$

Dengan

$\sum_{i=1}^n w_ie_i^2 = \sum_{i=1}^n w_i(Y_i - \beta_0X_i)$ 

$w_i = \frac{1}{\sigma_i}$

adanya *weight*($w_i$) atau pembobotan diperlukan untuk mendapatkan varians dari $e_i$ yang konstan.

$Var(e_i) \approx 1 \rightarrow Var(\epsilon) \approx \sigma^2$

Jika varians dari $e_i$ konstan, maka dapat di asumskan varians error nya mendekati $\sigma^2$.

sehingga $\hat{\beta_1}$ dari :

$\hat{\beta_1} = \frac{\sum X_iY_i - \sum X_i \sum Y_i}{\sum X_i^2 - (\sum X_i)^2}$

menjadi

$\hat{\beta_1} = \frac{(\sum w_i) (\sum w_i\sum X_iY_i) - (\sum w_i\sum X_i) (\sum w_i \sum Y_i)}{(\sum w_i)(\sum w_i\sum X_i^2) - (\sum w_i\sum X_i)^2}$ 

Var$(\hat{\beta_1}) = \frac{\sum w_i}{(\sum w_i)(\sum w_iX_i^2) - (\sum w_iX_i)^2}$

Contoh

$Y_i = \beta_0 + \beta_1X_1 + \epsilon_i$

$\epsilon_i = \beta_2X_2 + V_i$

Ini dapat menyebabkan heteroskedastisitas karena $\epsilon$ memiliki $X_2$ yang berkaitan dengan $Y_i$ sehingga diperlukan penanggulangan heteroskedastisitas.
